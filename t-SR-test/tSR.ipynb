```python
from sklearn import metrics
import math
```

```python
MAE = metrics.mean_absolute_error(data["F5"],data["Potential (v)"])
MSE = metrics.mean_squared_error(data["F5"],data["Potential (v)"])
MAPE = metrics.mean_absolute_percentage_error(data["F5"],data["Potential (v)"])
RMSE = math.sqrt(metrics.mean_squared_error(data["F5"],data["Potential (v)"]))
R2 = metrics.r2_score(data["F5"],data["Potential (v)"])

print('MAE:',MAE)
print('MSE:',MSE)
print('MAPE:',MAPE)
print('RMSE:',RMSE)
print('R2-score:',R2)
```
